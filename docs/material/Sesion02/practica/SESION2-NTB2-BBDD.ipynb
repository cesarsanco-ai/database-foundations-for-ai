## Actividad 3: Creación de una interfaz simple con Streamlit (Opcional)

**Herramientas**: Google Colab, Streamlit, ngrok.

Esta actividad muestra cómo crear una aplicación web simple que consulta un DataFrame y la exponemos públicamente con ngrok.

### 3.1. Crear la aplicación Streamlit

Ejecuta la siguiente celda para crear un archivo `app.py` con el código de la aplicación.

````python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np

# Título
st.title("📊 Dashboard de Ventas Simple")

# Crear datos de ejemplo
data = pd.DataFrame({
    'fecha': pd.date_range('2025-01-01', periods=30, freq='D'),
    'ventas': np.random.randint(50, 200, size=30),
    'categoria': np.random.choice(['Electrónica', 'Ropa', 'Hogar'], 30)
})

# Mostrar datos
st.subheader("Datos de ventas")
st.dataframe(data)

# Filtro por categoría
categoria = st.selectbox("Selecciona categoría", data['categoria'].unique())
filtrado = data[data['categoria'] == categoria]

# Gráfico de líneas
st.subheader(f"Ventas diarias - {categoria}")
st.line_chart(filtrado.set_index('fecha')['ventas'])

# Estadísticas
st.subheader("Estadísticas")
st.write(filtrado['ventas'].describe())

print("✅ Archivo app.py creado.")
````

### 3.2. Ejecutar Streamlit en segundo plano

Necesitamos ejecutar Streamlit en un proceso separado. Usaremos `threading` para ello.

````python
import threading
import subprocess
import time

def run_streamlit():
    subprocess.run(["streamlit", "run", "app.py", "--server.port", "8501", "--server.address", "0.0.0.0"])

# Iniciar Streamlit en un hilo separado
thread = threading.Thread(target=run_streamlit)
thread.start()

# Esperar a que Streamlit inicie
time.sleep(5)
print("✅ Streamlit iniciado en el puerto 8501")
````

### 3.3. Exponer la aplicación con ngrok

ngrok crea un túnel seguro a nuestra aplicación local, generando una URL pública.

````python
from pyngrok import ngrok

# Configurar ngrok (puede pedir autenticación gratuita en https://ngrok.com)
# Si es tu primera vez, necesitas registrarte y obtener un token
# ngrok.set_auth_token("TU_TOKEN")

# Crear túnel al puerto 8501
public_url = ngrok.connect(8501)
print(f"✅ Aplicación disponible en: {public_url}")
````

Abre la URL generada en cualquier navegador. Verás el dashboard interactivo de Streamlit.

**Nota**: La aplicación dejará de funcionar cuando detengas el kernel de Colab. La URL es pública mientras el túnel esté activo.
