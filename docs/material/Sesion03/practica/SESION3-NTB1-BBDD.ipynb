## Caso Práctico 1: Sistema de Biblioteca

**Enunciado:**

Se desea diseñar una base de datos para una biblioteca. Debe almacenar información sobre:
*   **Libros:** Cada libro tiene un ISBN (único), título, año de publicación y editorial. Un libro puede ser escrito por varios autores.
*   **Autores:** De cada autor se guarda un identificador único, nombre, nacionalidad y fecha de nacimiento.
*   **Socios:** De cada socio se guarda un número de socio único, nombre, dirección, teléfono y email.
*   **Préstamos:** Se debe registrar cada préstamo de un libro a un socio. Un préstamo tiene fecha de préstamo y fecha de devolución (puede ser nula si no se ha devuelto). Un socio puede tomar prestados varios libros a lo largo del tiempo, y un libro puede ser prestado múltiples veces (pero no simultáneamente). Asumimos que un préstamo es de un solo libro (para simplificar).

### Paso 1: Identificar Entidades, Atributos y Relaciones

**Entidades:**
*   **Libro** (fuerte)
*   **Autor** (fuerte)
*   **Socio** (fuerte)
*   **Préstamo** (débil? Depende de Libro y Socio, pero tiene sentido como fuerte con su propio ID)

**Atributos:**
*   **Libro:** ISBN (clave), título, año, editorial
*   **Autor:** id_autor (clave), nombre, nacionalidad, fecha_nac
*   **Socio:** id_socio (clave), nombre, dirección, teléfono, email
*   **Préstamo:** id_prestamo (clave), fecha_prestamo, fecha_devolucion

**Relaciones:**
*   **Escribe:** entre Autor y Libro. Un autor puede escribir varios libros, un libro puede tener varios autores. Es una relación **N:M** (muchos a muchos). No tiene atributos.
*   **Realiza:** entre Socio y Préstamo. Un socio puede realizar varios préstamos, un préstamo pertenece a un solo socio. Es una relación **1:N**.
*   **Contiene:** entre Préstamo y Libro. Un préstamo contiene un solo libro (por nuestra simplificación), un libro puede estar en varios préstamos (a lo largo del tiempo). Es una relación **1:N** desde Préstamo a Libro (varios préstamos pueden referirse al mismo libro).

### Paso 2: Crear el Diagrama E-R en draw.io

Ahora, usando draw.io, crea el diagrama Entidad-Relación. Debes incluir:
*   Rectángulos para las entidades.
*   Óvalos para los atributos (conectados a su entidad).
*   Rombo para la relación N:M (Escribe) y líneas para las relaciones 1:N.
*   Indicar las cardinalidades (1, N) en las líneas.

**Ejemplo de cómo podría verse (simplificado):**

```
    [Autor] ----< Escribe >---- [Libro]
       |                          |
       N                          N

    [Socio] ----< Realiza >---- [Préstamo] ----< Contiene >---- [Libro]
       1                          N              1              N
```

*(En el diagrama real, los atributos se conectan a cada entidad)*

### Paso 3: Convertir a Modelo Lógico (Tablas, PK, FK)

Aplicamos las reglas de transformación:

1.  **Entidades fuertes** se convierten en tablas.
2.  **Relaciones 1:N**: la clave de la entidad del lado "1" se añade como clave foránea (FK) en la tabla de la entidad del lado "N".
3.  **Relaciones N:M**: se crea una nueva tabla intermedia que contiene las claves primarias de ambas entidades como FK. La PK de esta tabla suele ser la combinación de ambas FK.

**Aplicado a nuestro caso:**

*   **Autor** (id_autor, nombre, nacionalidad, fecha_nac) – PK: id_autor
*   **Libro** (isbn, titulo, anio, editorial) – PK: isbn
*   **Socio** (id_socio, nombre, direccion, telefono, email) – PK: id_socio
*   **Préstamo** (id_prestamo, fecha_prestamo, fecha_devolucion, id_socio, isbn) – PK: id_prestamo; FK: id_socio, isbn.
    *Nota: La relación Contiene (1:N) se materializa con la FK `isbn` en Préstamo. La relación Realiza (1:N) se materializa con la FK `id_socio`.*
*   **Libro_Autor** (id_autor, isbn) – PK: (id_autor, isbn); FK: id_autor, isbn. (Tabla intermedia para la N:M).

**Escribe el modelo lógico en formato textual o en una celda de código como comentario:**

````python
# Modelo Lógico - Biblioteca
print("""
Tablas:
- Autor (id_autor, nombre, nacionalidad, fecha_nac)
- Libro (isbn, titulo, anio, editorial)
- Socio (id_socio, nombre, direccion, telefono, email)
- Prestamo (id_prestamo, fecha_prestamo, fecha_devolucion, id_socio, isbn)
- Libro_Autor (id_autor, isbn)
""")
````

## Reflexión y Conexión con Normalización

El modelo lógico que hemos construido para Ventas, ¿está en una forma normal adecuada? Por ejemplo:
*   **DetallePedido** tiene una clave compuesta (id_pedido, id_producto) y atributos no clave (cantidad, precio_unitario). ¿Dependen completamente de la clave? Sí, cantidad y precio_unitario dependen de la combinación (qué producto, en qué pedido).
*   En **Producto**, el atributo `nombre_categoria` no existe, porque hemos separado Categoría. Esto evita la dependencia transitiva (3FN).

## Conclusiones de la Sesión

*   El **modelo conceptual (E-R)** es una representación abstracta de los requisitos del negocio, independiente de la tecnología.
*   Las **entidades** se convierten en tablas, los **atributos** en columnas, y las **relaciones** se materializan mediante claves foráneas o tablas intermedias.
*   Las relaciones **1:N** se implementan añadiendo la PK del lado "1" como FK en la tabla del lado "N".
*   Las relaciones **N:M** requieren una tabla intermedia que contiene las PK de ambas entidades.
*   **draw.io** es una herramienta accesible y potente para crear diagramas E-R.
